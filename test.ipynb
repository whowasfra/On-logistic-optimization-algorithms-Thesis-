# Librerie
Si importano le seguenti librerie

```python
from py3dbp_adv import Item, Bin, Packer
```

Per l'analisi consideriamo i seguenti mezzi

```python
Ducato = Bin(name="Ducato",width=1.87,height=2.52,depth=6.0,max_weight=3000)
```

```python
packer = Packer()
```

```python
from py3dbp_adv import item_generator

orders = item_generator(
    w_range      = (0.05,1.5),
    h_range      = (0.05,1.5),
    d_range      = (0.05,1.5),
    weight_range = (0.01,150),
    priority_range = (0,10),
    batch_size   = 600
)
```

```python
packer.set_default_bin(Ducato)
packer.add_batch(orders)
```

```python
packer.pack()
```

```python
print(len(packer.items))
print(len(packer.unfit_items))

for bin in packer.current_configuration:
    print(len(bin.unfitted_items))
    bin.render()
```

```python
print(len(packer.current_configuration))
print(packer.calculate_statistics())
#print(packer.statistics["loaded_volume"]/packer.)
```
